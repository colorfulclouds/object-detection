```python
#refer:https://blog.csdn.net/two_vv/article/details/76769860
#alexnet原始模型以及预训练参数导入

class AlexNet_model(object):
    def __init__(self , x , keep_prob , num_classes , skip_layer):
        self.X = x
        self.NUM_CLASSES = num_classes
        self.KEEP_PROB = keep_prob
        self.SKIP_LAYER = skip_layer #需要训练的层 不使用预训练参数
        self.WEIGHTS_PATH = PRETRAIN_ALEXNET_MODEL_PATH
                
        #创建fc8
        self.model()
    
    def fine_tuning(self):
        pass
    
    #构建alexnet模型
    def model(self):
        conv1 = self.conv(self.X ,  11 ,11 , 96 , 4 , 4 , name='conv1' , padding='VALID')
        pool1 = self.max_pool(conv1 , 3,3,2,2,padding='VALID' , name='pool1')
        norm1 = self.lrn(pool1 , 2 , alpha=2e-5 , beta=0.75 , name='norm1')
        
        conv2 = self.conv(norm1 , 5,5,256,1,1,groups=2,name='conv2')
        pool2 = self.max_pool(conv2 , 3,3,2,2,padding='VALID' , name='pool2')
        norm2 = self.lrn(pool2 , 2 , alpha=2e-5 , beta=0.75 , name='norm2')
        
        conv3 = self.conv(norm2 , 3,3,384,1,1,name='conv3')
        
        conv4 = self.conv(conv3 , 3,3,384,1,1,groups=2 , name='conv4')
        
        conv5 = self.conv(conv4 , 3,3,256,1,1,groups=2 , name='conv5')
        pool5 = self.max_pool(conv5 , 3,3,2,2,padding='VALID' , name='pool5')
        
        #FC层都是需要训练的
        #FC层捕获高层特征
        #conv层捕获低层特征
        flattened = tf.reshape(pool5 , shape=(-1 , 6*6*256))
        fc6 = self.fc(flattened , 6*6*256 , 4096 , name='fc6')
        dropout6 = self.dropout(fc6 , self.KEEP_PROB)
        
        self.fc7 = self.fc(dropout6 , 4096 , 4096 , name='fc7') #这一层作为4096维特征使用
        dropout7 = self.dropout(self.fc7 , self.KEEP_PROB)
        
        self.fc8 = self.fc(dropout7 , 4096 , self.NUM_CLASSES , relu=False , name='fc8') #模型的输出
        
#ValueError: Shapes must be equal rank, but are 2 and 1 for 'fc6_1/Assign_1' (op: 'Assign') with input shapes: [9216,4096], [4096]. 
    #装载预训练参数
    def load_parameter(self , session): #session回话使用
        weight_dict = np.load(self.WEIGHTS_PATH , encoding='bytes').item()
        
        for op_name in weight_dict:
            if op_name not in self.SKIP_LAYER:
                #使用预训练参数
                with tf.variable_scope(op_name , reuse=True):
                    for data in weight_dict[op_name]:
                        if len(data.shape) == 1:
                            #biases
                            var = tf.get_variable('biases' , trainable=False)
                            session.run(var.assign(data))
                        else:
                            #weights
                            var = tf.get_variable('weights' , trainable=False)
                            session.run(var.assign(data))

    #功能函数
    def conv(self , x , filter_height , filter_width , filter_num , stride_y , stride_x , name , padding='SAME' , groups = 1):
        input_channels = int(x.get_shape()[-1]) #image channel
        
        def conv_builtin(i , k):
            return tf.nn.conv2d(i , k , strides=[1 , stride_y , stride_x , 1] , padding=padding)

        with tf.variable_scope(name) as scope:
            weights = tf.get_variable('weights' , shape=[filter_height , filter_width , input_channels/groups , filter_num])
            
            biases = tf.get_variable('biases' , shape=[filter_num])
            
            #两块显卡需要分开
            if groups == 1:
                conv = conv_builtin(x , weights)
            else:
                input_groups = tf.split(value=x , num_or_size_splits=groups , axis=3)
                weight_groups = tf.split(value=weights , num_or_size_splits=groups , axis=3)
                output_groups = [conv_builtin(i , k) for i,k in zip(input_groups , weight_groups)] #zip压缩函数 zip(*xxx)解压缩函数
                
                conv = tf.concat(values=output_groups , axis=3)
            
            bias = tf.reshape(tf.nn.bias_add(conv , biases) , shape=[-1] + conv.get_shape().as_list()[1:]) #卷积计算后 #-1有tensorflow自行维度匹配
            
            relu = tf.nn.relu(bias , name=scope.name) #卷积计算后进行激活
            
            return relu
      
    def max_pool(self , x , filter_height , filter_width , stride_y , stride_x , name , padding='SAME'):
        return tf.nn.max_pool(x , ksize=[1,filter_height , filter_width , 1] ,
                              strides=[1 , stride_y , stride_x , 1] , padding=padding , name=name)
    
    def lrn(self , x , radius , alpha , beta , name , bias=1.0):
        return tf.nn.local_response_normalization(x , depth_radius=radius , alpha=alpha , beta=beta , bias=bias , name=name)
    
    def dropout(self , x , keep_prob):
        return tf.nn.dropout(x , keep_prob=keep_prob)
    
    def fc(self , x , input_size , output_size , name , relu = True):
        with tf.variable_scope(name) as scope:
            weights = tf.get_variable('weights' , shape=[input_size , output_size] , trainable=True)
            biases = tf.get_variable('biases' , shape=[output_size] , trainable=True)
            
            activation = tf.nn.xw_plus_b(x , weights , biases , name=scope.name)
            
            if relu == True:
                relu = tf.nn.relu(activation)
                
                return relu
            else:
                return activation
    
```

```python
class Xml_process(object):
    def __init__(self):
        self.xml_file_names_train = glob(TRAIN_XML_PATH + '*') #所有的xml文件
        #self.xml_file_names_test = glob(TEST_XML_PATH + '*') #测试数据集不用解析 直接生成结果
        
    #读一个xml文件
    #参数为xml文件的路径
    def meta_analysis(self , xml_file):
        xml_file = xml.dom.minidom.parse(xml_file)
        xml_file_docu_ele = xml_file.documentElement
        
        filename_list = xml_file_docu_ele.getElementsByTagName('filename')
        #filename_list可能有多个filename的 所以要索引0(此数据集中filename只有一个)
        filename = filename_list[0].childNodes[0].data #filename_list.firstChild.data
        
        #图像的尺寸信息
        size_list = xml_file_docu_ele.getElementsByTagName('size')

        for size in size_list:
            width_list = size.getElementsByTagName('width')
            width = int(width_list[0].childNodes[0].data)

            height_list = size.getElementsByTagName('height')
            height = int(height_list[0].childNodes[0].data)

            channel_list = size.getElementsByTagName('depth')
            channel = int(channel_list[0].childNodes[0].data)
        
        shape = (width , height , channel)
        
        #一个文件中有多个object
        object_list = xml_file_docu_ele.getElementsByTagName('object')
        
        #多个object与多个object对应的详细信息
        name_boxes = [] #一个元素就是一个object
        crop_boxes = []

        for objects in object_list:
            #一次循环处理一个object信息
            #一个xml文件（即一个图像中）有多个object

            name_list = objects.getElementsByTagName('name')

            #name_box中第0个元素是object的名称 后面的是详细物体part的名称
            name_box = []

            for i in range(len(name_list)):
                name_box.append(name_list[i].childNodes[0].data)

            bndbox = objects.getElementsByTagName('bndbox')

            #crop_box中第0个元素是object的坐标 后面的为详细物体part的坐标
            crop_box = []

            for box in bndbox:
                
                int(round(float('1.0542')))
                x1_list = box.getElementsByTagName('xmin')
                x1 = int( round( float(x1_list[0].childNodes[0].data) ) )

                y1_list = box.getElementsByTagName('ymin')
                y1 = int(round(float( y1_list[0].childNodes[0].data )))

                x2_list = box.getElementsByTagName('xmax')
                x2 = int(round(float( x2_list[0].childNodes[0].data )))

                y2_list = box.getElementsByTagName('ymax')
                y2 = int(round(float( y2_list[0].childNodes[0].data )))

                crop_box.append([x1,x2,y1,y2])

            name_boxes.append(name_box)
            crop_boxes.append(crop_box)
            
        return shape , name_boxes , crop_boxes

    #将所有的xml读完
    def analysis(self):
        img_property = []
        
        for i in range(len(self.xml_file_names_train)):
            shape , name_boxes , crop_boxes = self.meta_analysis(self.xml_file_names_train[i])
            
            img_property.append([shape , name_boxes , crop_boxes])
            
        return img_property #一个元素为一个图像的全部信息


#test = Xml_process()
#test.meta_analysis(test.xml_file_names_train[10])
```

```python
#图像预处理
class Process_img(object):
    def __init__(self):
        self.height = HEIGHT
        self.width = WIDTH
        self.channel = CHANNEL
        
        self.batch_size = BATCH_SIZE
        
        self.img_shape = IMG_SHAPE
        
        self.img_file_names_train = glob(TRAIN_DATA_PATH+'*') #训练全路径信息
        self.img_file_names_test = glob(TEST_DATA_PATH+'*') #测试全路径信息
        
        #self.img_num = len(self.img_file_names_train) #图像数量
        
        #图像信息xml读出模块
        self.xml_m = Xml_process()
        
    def __meta_load(self , img_name):
        #img_arr = imageio.imread(img_name) #使用此函数打开 导致迁移失败 RGB height*width*channel
        img_arr = cv2.imread(img_name) #BGR height*width*chanel
        #img_arr = np.array(img_arr)
        
        xml_file_name = TRAIN_XML_PATH + img_name[-15:-4] +  '.xml'
        
        shape , name_boxes , crop_boxes = self.xml_m.meta_analysis(xml_file_name)
        
        proposal = [] #存储bndbox的图像数据信息

        for i in range(len(crop_boxes)): #多个object
            an_object = []
            
            #j循环注意len(crop_boxes[i]) crop_boxes不是标准矩形形状
            for j in range(len(crop_boxes[i])): #一个object中可能有part部分 也可能没有 但第一个元素一定是object的bndbox信息
                x1 = crop_boxes[i][j][0]
                x2 = crop_boxes[i][j][1]
                y1 = crop_boxes[i][j][2]
                y2 = crop_boxes[i][j][3]
                
                an_object.append(img_arr[y1:y2 , x1:x2 , :]) #正确的切片格式
                name_boxes[i][j] = STR2LABEL.get(name_boxes[i][j] , 'none') #将str的label数字化 none用来防止part部分 暂时不考虑part部分
                
            proposal.append(an_object)
            
        return img_arr , shape , proposal , name_boxes , crop_boxes #shape暂时没用
    
    def load(self):
        #
        batch_img = np.random.choice(self.img_file_names_train , size=self.batch_size) #随机文件路径
        #idx = np.random.randint(0 , high = len(self.img_file_names_train) , size = self.batch_size) #随机索引
        
        imgs = []
        imgs_shape = [] #图像的尺寸信息
        proposals = [] #候选区域具体图像数据 shape: batch_size*proposal*height*width*channel
        labels = [] #候选区域的人工标记 shape: batch_size*proposal*1 #如果使用softmax分类 需要进行one-hot化
        G_box = [] #候选区域形式 (x,y,w,h) shape: batch_size*proposal*4 #4为(x,y,w,h)
        img_names = [] #图片的名字
        
        for i in batch_img:
            img_names.append(i[-15:-4])
            
            img_arr , _ , proposal , name_boxes , crop_boxes = self.__meta_load(i) #_为图像的尺寸信息 暂时用不上
            
            imgs.append(img_arr)
            
            #imgs_shape.append(_)
            
            proposals.append(proposal)
            labels.append(name_boxes)
            G_box.append(crop_boxes)
        
        #原图像 候选区域具体数据 候选区域的人工标记 候选区域的位置信息 文件名        
        return np.array(imgs) , proposals , labels , G_box , img_names

    #def load_test(self):
    #    batch_img = np.random.choice(self.img_file_names_test , size=self.batch_size)
    #    #idx = np.random.randint(0 , high = len(self.img_file_names_test) , size = self.batch_size)
    #        
    #    imgs = []
    #    
    #    for i in batch_img:
    #        img_arr = imageio.imread(i)
    #        img_arr = transform.imresize(img_arr , (self.height , self.width)) #224 224 3
    #        imgs.append(img_arr)
    #    
    #    return np.array(imgs) #先不除以255 在候选区域产生之后再考虑 shape: batch_size*height*width*channel

    #svm使用
    #bbox回归使用
    def get_crop_boxes(self , path):
        xml_file_name = TRAIN_XML_PATH + path[-15:-4] +  '.xml'
        
        return  self.xml_m.meta_analysis(xml_file_name)
        
```

```python
#候选区域的产生
class Clip_img(object):
    def __init__(self):
        self.size_threshold = 220 #ss算法产生过小的bbox略去

    #单张图像裁剪函数 产生2k-3k的候选区域
    #返回值：候选区域的位置信息
    def __meta_clip(self , meta_img): #meta_img shape: height*width*channel
        #关键函数
        meta_img_with_labels , regions = ss.selective_search(meta_img , 500 , 0.9 , 10) #可以调整ss算法的参数
        
        rect = []
        labels = [] #
        size = [] #
        
        no_repeat = set() #保证候选框不重复出现
        
        for i in range(len(regions)):
            
            if regions[i]['rect'] in no_repeat: #已经存在了
                continue
            
            #bbox size较小的略去
            if regions[i]['size'] < self.size_threshold:
                continue
            
            x = regions[i]['rect'][0]
            y = regions[i]['rect'][1]
            w = regions[i]['rect'][2]
            h = regions[i]['rect'][3]
            
            #存在长或宽为0的bbox略去
            if w==0 or h==0:
                continue
            
            d0 , d1 , d2 = meta_img[y:y+h , x:x+w , :].shape #正确的切片格式
            if d0==0 or d1==0 or d2==0:
                continue
                
            no_repeat.add(regions[i]['rect']) #添加到集合中
            
            rect.append(np.array([x,x+w,y,y+h])) #每一个元素是一个候选区域的框
            #labels.append(str(regions[i]['labels'][0])) #每一个元素是一个label
            #size.append(regions[i]['size'])

        return np.array(rect) #shape: proposal*4 (x1,x2,y1,y2)
        
    #产生候选区域的位置信息
    def clip(self , img):
        img_clip = []
        
        for i in range(img.shape[0]): #batch size
            img_clip.append(self.__meta_clip(img[i]))
        
        return np.array(img_clip) #候选区域位置信息 shape: batch_size*proposal*4
    
    #返回数组表示候选区域的具体数据信息
    def __meta_clip_region(self , meta_img , meta_proposal_position):
        region_meta_img = []
        
        for i in range(meta_proposal_position.shape[0]):
            x1 = meta_proposal_position[i][0]
            x2 = meta_proposal_position[i][1]
            y1 = meta_proposal_position[i][2]
            y2 = meta_proposal_position[i][3]
            
            clip_region = meta_img[y1:y2 , x1:x2 , :] #正确的切片格式
            region_meta_img.append(clip_region)
        
        return np.array(region_meta_img) #shape: proposal*height*width*3
    
    def clip_region(self , imgs , proposal_position): #batch_size height width channel
        
        img_clip_region = []
        
        for i in range(imgs.shape[0]): #batch size
            img_clip_region.append( self.__meta_clip_region(imgs[i] , proposal_position[i]) )
        
        return np.array(img_clip_region) #候选区域位置信息 shape: batch_size*proposal*height*width*3

    
    #推理阶段使用 对外暴露接口
    def meta_clip(self , meta_img):
        return self.__meta_clip(meta_img)
    
    def meta_clip_region(self , meta_img , meta_proposal_position):
        region_meta_img = []
        
        for i in range(len(meta_proposal_position)):
            x1 = meta_proposal_position[i][0]
            x2 = meta_proposal_position[i][1]
            y1 = meta_proposal_position[i][2]
            y2 = meta_proposal_position[i][3]
            
            clip_region = meta_img[y1:y2 , x1:x2 , :] #正确的切片格式
            
            clip_region_resize = cv2.resize(clip_region , dsize=(HEIGHT , WIDTH)) #resize合适的大小
            
            region_meta_img.append(clip_region_resize)
        
        return np.array(region_meta_img) #shape: proposal*227*227*3
```

```python
class Img_generator(object):
    def __init__(self):
        
        self.pr_generator = Clip_img()
        self.img_loader = Process_img()

    #计算bbox面积
    def bbox_area(self , bbox):
        w = bbox[1] - bbox[0]
        h = bbox[3] - bbox[2]
        
        return w*h
    
    #计算交并比
    def IoU(self , bbox_a , bbox_b):
        xmin_a = bbox_a[0]
        xmax_a = bbox_a[1]
        ymin_a = bbox_a[2]
        ymax_a = bbox_a[3]
        
        xmin_b = bbox_b[0]
        xmax_b = bbox_b[1]
        ymin_b = bbox_b[2]
        ymax_b = bbox_b[3]
        
        if xmin_a < xmax_b <= xmax_a and (ymin_a < ymax_b <= ymax_a or ymin_a <= ymin_b < ymax_a):
            flag = True
        elif xmin_a <= xmin_b < xmax_a and (ymin_a < ymax_b <= ymax_a or ymin_a <= ymin_b < ymax_a):
            flag = True
        elif xmin_b < xmax_a <= xmax_b and (ymin_b < ymax_a <= ymax_b or ymin_b <= ymin_a < ymax_b):
            flag = True
        elif xmin_b <= xmin_a < xmax_b and (ymin_b < ymax_a <= ymax_b or ymin_b <= ymin_a < ymax_b):
            flag = True
        else:
            flag = False
        
        if flag:
            x_sorted_list = sorted([xmin_a, xmax_a, xmin_b, xmax_b])
            y_sorted_list = sorted([ymin_a, ymax_a, ymin_b, ymax_b])
            x_intersect_w = x_sorted_list[2] - x_sorted_list[1]
            y_intersect_h = y_sorted_list[2] - y_sorted_list[1]
            area_inter = x_intersect_w * y_intersect_h
            
            union_area = self.bbox_area(bbox_a) + self.bbox_area(bbox_b) - area_inter
            
            return area_inter/union_area
        else:
            return 0.0
    
    def __meta_train_proposal(self , label , G_box , proposal_position , proposal_value , svm=False):
        #====
        positive_labels = []
        negative_labels = []
        
        positive_data = []
        negative_data = []
        
        positive_G_box = []
        negative_G_box = []
        #====
        
        for i in range(len(G_box)): #一幅图的人工标定bbox proposal
            for j in range(len(proposal_position)): #一幅图的ss产生的bbox
                
                resize_proposal_value_j = cv2.resize(proposal_value[j] , dsize=(HEIGHT , WIDTH))
                
                if self.IoU(G_box[i][0] , proposal_position[j]) < IoU_THRESHOLD: #0.5
                    #小于阈值 label为背景
                    negative_labels.append(0)
                    negative_G_box.append(G_box[i][0])
                    
                    negative_data.append(resize_proposal_value_j)
                    
                else:
                    #大于阈值 label为正常label
                    positive_labels.append(label[i][0])
                    positive_G_box.append(G_box[i][0]) #bbox回归使用 当label不是背景的时候 才进行bbox回归
                    
                    positive_data.append(resize_proposal_value_j)
                
        return (np.array(positive_data) , to_categorical(positive_labels , num_classes=CLASSES_NUM) , positive_G_box,
                 np.array(negative_data) , to_categorical(negative_labels , num_classes=CLASSES_NUM) , negative_G_box)
    
    def get_train_proposal(self , imgs , labels , G_box , svm=False):
        proposal_position = self.pr_generator.clip(imgs) #候选区域由ss算法产生的bbox batch_size*proposal*4
        proposal_value = self.pr_generator.clip_region(imgs , proposal_position) #候选区域的具体图像数据 batch_size*proposal_height*width*3
        
        positive_train_proposal_data = [] #训练data batch_size*(G_box*proposal)*227*227*3
        positive_train_proposal_labels = [] #训练label batch_size*(G_box*proposal)*(1 or CLASSES_NUM)
        positive_train_proposal_G_box = [] #bbox回归使用 batch_size*(G_box*proposal)*4
        
        negative_train_proposal_data = []
        negative_train_proposal_labels = []
        negative_train_proposal_G_box = []
        
        for i in range(imgs.shape[0]): #batch size
            positive_data , positive_label , positive_G_box_c , negative_data , negative_label , negative_G_box_c = self.__meta_train_proposal(labels[i], G_box[i], proposal_position[i], proposal_value[i], svm)
            
            positive_train_proposal_data.append(positive_data)
            positive_train_proposal_labels.append(positive_label)
            positive_train_proposal_G_box.append(positive_G_box_c)
            
            negative_train_proposal_data.append(negative_data)
            negative_train_proposal_labels.append(negative_label)
            negative_train_proposal_G_box.append(negative_G_box_c)
            
        return (np.array(positive_train_proposal_data) , positive_train_proposal_labels , positive_train_proposal_G_box,
                 np.array(negative_train_proposal_data) , negative_train_proposal_labels , negative_train_proposal_G_box)
            
    def load(self , svm=False):
        imgs , _ , labels , G_box , img_names = self.img_loader.load() #第二个参数为人工标定框区域具体数据

        return self.get_train_proposal(imgs , labels , G_box , svm)

    #推理阶段使用
    #Alexnet_finetune demo 和 SVM_set中使用
    def one_img_rect_region(self , path): #返回一张图像的 数据 候选框坐标 候选区域详细数据
        img_arr = cv2.imread(path) #BGR height*width*chanel
        
        rects = self.pr_generator.meta_clip(img_arr) #proposal*4(x1,x2,y1,y2) selectivesearch算法产生
        rects_region = self.pr_generator.meta_clip_region(img_arr , rects) #proposal*227*227*3
        
        return img_arr , rects , rects_region
    
    #svm使用
    def one_img_rect_region_label_svm(self , path , label): ##返回一张图像的 数据 候选框坐标 候选区域详细数据 label
        img_arr = cv2.imread(path) #BGR height*width*chanel
        
        rects = self.pr_generator.meta_clip(img_arr) #proposal*4(x1,x2,y1,y2) #selectivesearch算法产生的候选框
        
        rects_region = self.pr_generator.meta_clip_region(img_arr , rects) #proposal*227*227*3
        
        labels = [] #标记
        G_boxes = [] #人工标定的框
        regions = [] #候选区域数据
        boxes = [] #候选框 #后续可能需要返回
        
        #获取人工标定的框
        shape , name_boxes , crop_boxes = self.img_loader.get_crop_boxes(path)
        
        for i in range(len(name_boxes)):
            if STR2LABEL[name_boxes[i][0]] == label:
                G_boxes.append(crop_boxes[i][0])
        
        #svm训练负样本
        for i in range(len(G_boxes)): #一幅图的与label相对应的人工标定bbox proposal
            for j in range(len(rects)): #一幅图的ss产生的bbox
                
                if self.IoU(G_boxes[i] , rects[j]) < SVM_IoU_THRESHOLD: #0.3
                    labels.append(0)
                
                    boxes.append(rects[j])
                    regions.append(rects_region[j])
        
        #svm训练正样本
        #与label对应的人工标定框的图像数据
        G_boxes_region = self.pr_generator.meta_clip_region(img_arr , G_boxes) #proposal*227*227*3
        
        for i in range(len(G_boxes)):
            labels.append(label)
            boxes.append(G_boxes[i])
            regions.append(G_boxes_region[i])
                
        return np.array(regions) , labels , np.array(boxes) , np.array(G_boxes) #后续可能需要返回boxes
    
    #bbox回归使用
    def __to_t(self , G_box , P_box):
        def to(rect):
            x1 = rect[0]
            x2 = rect[1]
            y1 = rect[2]
            y2 = rect[3]
            
            w = x2-x1
            h = y2-y1
            
            x_c = (x1+x2)//2
            y_c = (y1+y2)//2
            
            return x_c , y_c , w , h
        
        G_x , G_y , G_w , G_h = to(G_box)
        P_x , P_y , P_w , P_h = to(P_box)
        
        t_x = (G_x-P_x)//P_w
        t_y = (G_y-P_y)//P_h
        t_w = np.log(G_w/P_w)
        t_h = np.log(G_h/P_h)
        
        return t_x , t_y , t_w , t_h
    
    def one_img_rect_region_label_bnd(self , path , label):
        img_arr = cv2.imread(path) #BGR height*width*chanel
        
        rects = self.pr_generator.meta_clip(img_arr) #proposal*4(x1,x2,y1,y2)
        
        rects_region = self.pr_generator.meta_clip_region(img_arr , rects) #proposal*227*227*3
        
        G_boxes = [] #人工标定的框
        regions = [] #候选区域数据
        targets = [] #bbox回归使用的训练数据中的y
        
        #获取人工标定的框
        shape , name_boxes , crop_boxes = self.img_loader.get_crop_boxes(path)
        
        for i in range(len(name_boxes)):
            if STR2LABEL[name_boxes[i][0]] == label:
                G_boxes.append(crop_boxes[i][0])
            
        for i in range(len(G_boxes)): #一幅图的人工标定bbox proposal
            for j in range(len(rects)): #一幅图的ss产生的bbox
                
                if self.IoU(G_boxes[i] , rects[j]) > BBOX_REGRESS_IoU_THRESHOLD:
                    target = self.__to_t(G_boxes[i] , rects[j])
                    
                    targets.append(target)
                    regions.append(rects_region[j])
                
        return np.array(regions) , np.array(targets) #区域具体数据和t
```

```python
#refer:https://blog.csdn.net/two_vv/article/details/76769860

class Alexnet_finetune(object):
    def __init__(self):
        #迁移参数
        self.learning_rate = 0.001
        self.num_epochs = 1 #一个epoch需要12个小时 crazy!!!
        self.xxx = 18000 #PASCAL VOC所有的数据
        self.dropout_rate = 0.5
        self.train_layers = ['fc7' , 'fc8'] #只训练fc7 fc8层 别的层使用预训练参数
        self.classes_num = CLASSES_NUM #20+1
        
        #生成训练数据使用
        self.img_generator = Img_generator()

        #模型可视化 模型保存相关
        #==========
        self.filewriter_path = 'qp/image' #tensorboard使用
        self.checkpoint_path = 'qp/finetune_alexnet' #模型及其参数的保存 持久化
        #==========
        
        #下面是tensorflow部分
        self.x = tf.placeholder(tf.float32 , shape=[None , HEIGHT , WIDTH , CHANNEL])
        self.y = tf.placeholder(tf.float32 , shape=[None , self.classes_num]) #None 行不定 tf自行完成判定
        self.keep_prob = tf.placeholder(tf.float32) #scalar
        
        self.model = AlexNet_model(self.x , self.keep_prob , self.classes_num , self.train_layers)

        self.score = self.model.fc8
        
        #预训练结束之后 提取4096维特征使用
        self.features = self.model.fc7 #fc7 为4096维输出
    
        #需要自己训练的层参数
        #fc7 weights biases
        #fc8 weights biases
        self.var_list = [v for v in tf.trainable_variables() if v.name.split('/')[0] in self.train_layers]
        
        #损失函数 交叉熵 softmax
        with tf.name_scope('cross_ent'):
            self.loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=self.score , labels=self.y))
        
        #构造训练过程
        with tf.name_scope('train'):
            #损失对fc7 weights biases fc8 weights biases的梯度
            self.gradients = tf.gradients(self.loss , self.var_list) #损失对训练参数的梯度
            self.gradients = list(zip(self.gradients , self.var_list))
            
            #没有使用minimize 而是使用appl_gradients方式 梯度下降
            self.optimizer = tf.train.GradientDescentOptimizer(self.learning_rate)
            self.train_op = self.optimizer.apply_gradients(grads_and_vars = self.gradients)
            
        #可视化使用
        for gradient , var in self.gradients:
            tf.summary.histogram(var.name + '/gradient' , gradient)
        
        for var in self.var_list:
            tf.summary.histogram(var.name , var)
            
        tf.summary.scalar('cross_entropy' , self.loss)
        
        #精度
        with tf.name_scope('accuracy'):
            self.corrent_pred = tf.equal(tf.argmax(self.score , axis=1) , tf.argmax(self.y , axis=1)) #返回的是最大值的索引
            self.accuracy = tf.reduce_mean(tf.cast(self.corrent_pred , tf.float32))
        
        tf.summary.scalar('accuracy' , self.accuracy)
        
        
        #merge所有summary 使用tensorboard显示
        self.merged_summary = tf.summary.merge_all() #需要有session时才能起作用
        #tensorboard使用 日志形式 在此文件夹下运行tensorboard
        self.writer = tf.summary.FileWriter(logdir = self.filewriter_path ) #qp/image
        #将模型保存起来 持久模式 以后能够读出使用
        self.saver = tf.train.Saver()
        
        #防止显存溢出添加
        self.resource_threshold_shape_0 = 800 #候选框超过800 将其拆分为两块处理 #1000还是有点大

    def train(self):
        
        #BATCH_SIZE 建议设置为1
        def next_batch():
            try:
                positive_train_data , positive_train_labels , _ , negative_train_data , negative_train_labels , _ = self.img_generator.load() #载入batch_size的图像
            except ValueError:
                #ValueError: could not broadcast input array from shape (188,128,3) into shape (188)
                #没找到原因 重新load一张图
                #2008_006489.jpg 出现问题
                positive_train_data , positive_train_labels , _ , negative_train_data , negative_train_labels , _ = next_batch()
            
            return positive_train_data , positive_train_labels , negative_train_data , negative_train_labels , len(positive_train_data[0]) + len(negative_train_data[0]) #最后一个为proposal数量
        
        def next_proposal(positive_train_data , positive_train_labels , negative_train_data , negative_train_labels):
            #32正样本 96负样本
            positive_train_data_batch = []
            positive_train_labels_batch = []
            
            negative_train_data_batch = []
            negative_train_labels_batch = []
            
            #paper中正负样本比例
            if len(positive_train_data[0]) == 0:
                #正样本为0
                #全部使用负样本
                negative_proposal_idx = np.random.randint(0 , high=len(negative_train_data[0]) , size=PROPOSAL_SIZE) #128
                
                for k in negative_proposal_idx:
                    negative_train_data_batch.append(negative_train_data[0][k])
                    negative_train_labels_batch.append(negative_train_labels[0][k])
                
                return np.array(negative_train_data_batch) , np.array(negative_train_labels_batch)
            else:
                positive_proposal_idx = np.random.randint(0 , high=len(positive_train_data[0]) , size=PROPOSAL_SIZE_POSITIVE) #32
            
            if len(negative_train_data[0]) == 0:
                #负样本为0
                #全部使用正样本
                positive_proposal_idx = np.random.randint(0 , high=len(positive_train_data[0]) , size=PROPOSAL_SIZE) #128
                
                for j in positive_proposal_idx:
                    positive_train_data_batch.append(positive_train_data[0][j])
                    positive_train_labels_batch.append(positive_train_labels[0][j])
                    
                return np.array(positive_train_data_batch) , np.array(positive_train_labels_batch)
            else:
                negative_proposal_idx = np.random.randint(0 , high=len(negative_train_data[0]) , size=PROPOSAL_SIZE_NEGATIVE) #96
            
            #=====================
            #=====================
            
            for j in positive_proposal_idx:
                positive_train_data_batch.append(positive_train_data[0][j])
                positive_train_labels_batch.append(positive_train_labels[0][j])

            for k in negative_proposal_idx:
                    negative_train_data_batch.append(negative_train_data[0][k])
                    negative_train_labels_batch.append(negative_train_labels[0][k])
            
            positive_train_data_batch = np.array(positive_train_data_batch)
            positive_train_labels_batch = np.array(positive_train_labels_batch)
            
            negative_train_data_batch = np.array(negative_train_data_batch) 
            negative_train_labels_batch = np.array(negative_train_labels_batch) 
            
            return np.concatenate((positive_train_data_batch , negative_train_data_batch)) , np.concatenate((positive_train_labels_batch , negative_train_labels_batch))
        
        with tf.Session() as sess:
            
            if os.path.exists('../RCNN/qp/finetune_alexnet/checkpoint'):
                sess.run(tf.global_variables_initializer())
                self.saver.restore(sess , tf.train.latest_checkpoint('../RCNN/qp/finetune_alexnet/')) #读取模型
                
            else:
                sess.run(tf.global_variables_initializer())

                #计算图放到tensorboard中
                self.writer.add_graph(sess.graph)
                self.model.load_parameter(sess)
            
            for i in range(self.num_epochs): #10
                #一个epoch
                
                for k in range(self.xxx): #18000 所有图片
                    
                    positive_train_data , positive_train_labels , negative_train_data , negative_train_labels , train_proposal_num = next_batch() #选择一张图片

                    #将一张图像中的proposal全部训练一遍
                    iters = train_proposal_num // PROPOSAL_SIZE #训练完所有的proposal需要迭代多少次
                    for j in range(iters):
                        train_images_batch , train_labels_batch = next_proposal(positive_train_data , positive_train_labels , negative_train_data , negative_train_labels) #在选中的一张图片中选择一些候选框
                        #print(train_images_batch.shape , train_labels_batch.shape) #debug

                        losss , _ = sess.run([self.loss , self.train_op] , feed_dict={self.x:train_images_batch , self.y:train_labels_batch , self.keep_prob:self.dropout_rate })

                        #save model
                        #生成训练时的相关数据 上面使用tf.summary.xxx函数的数据
                        
                        #放在这里与放在for-j循环外面是一样的 但是就不能使用train_images_batch train_labels_batch了
                        if j==iters-1 and k%100 == 0: #100张图像写入一次
                            merge_summary = sess.run(self.merged_summary , {self.x:train_images_batch , self.y:train_labels_batch , self.keep_prob:1.0 })
                            self.writer.add_summary(merge_summary , i) #写入文件

                    print('epoch:%d/%d img:%d/%d loss:%f' % (i , self.num_epochs , k , self.xxx , losss))

                    if k % 500 == 0: #500张图像保存一次模型
                        checkpoint_name = os.path.join(self.checkpoint_path , 'model_epoch.ckpt')
                        save_path = self.saver.save(sess , checkpoint_name)
                
        self.writer.close()
        
    #使用调优后的alexnet进行特征提取
    def extract_feature(self , regions):
        
        #没有model的保存数据就先训练
        if not os.path.exists('../RCNN/qp/finetune_alexnet/checkpoint'):
            self.train()
            
        with tf.Session() as sess:
            sess.run(tf.global_variables_initializer())
            self.saver.restore(sess , tf.train.latest_checkpoint('../RCNN/qp/finetune_alexnet/')) #读取模型整个model的模型参数

            if regions.shape[0]>self.resource_threshold_shape_0:
                regions_cnn_features = []
                #拆开获取4096维特征
                for idx in range(regions.shape[0] // self.resource_threshold_shape_0):
                    regions_cnn_features_block = sess.run(self.features , feed_dict={self.x : regions[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0] , self.keep_prob:1.0})
                    regions_cnn_features.append(regions_cnn_features_block)

                #余下的部分（不是整的）
                regions_cnn_features_extra = sess.run(self.features , feed_dict={self.x : regions[(idx+1)*self.resource_threshold_shape_0:] , self.keep_prob:1.0})
                regions_cnn_features.append(regions_cnn_features_extra)
                
                regions_cnn_features = np.concatenate(regions_cnn_features) #将块的4096维特征合起来

            else:
                regions_cnn_features = sess.run(self.features , feed_dict={self.x : regions , self.keep_prob:1.0})
            
            return regions_cnn_features

    def demo(self , path):
        #
        data , rects , rects_region = self.img_generator.one_img_rect_region(path)
        
        
        return self.extract_feature(rects_region) , rects , rects_region
```

```python
#读取与特定label相关的图像数据并产生训练SVM使用的数据集
class Object_load(object):
    def __init__(self):
        self.object_path = [OBJECT_PATH+LABEL2STR[i]+'_train.txt' for i in range(1 , len(LABEL2STR))]
        self.img_generator = Img_generator()
    
    
    def load(self , label): #label为load的数据类型 1-20
        imgs_path = []
        
        for line in open(self.object_path[label-1]).readlines(): #需要减1 list下标从0开始
            line = line.rstrip('\n') #去掉末尾的\n
            
            line_split = line.split(' ')
            
            if line_split[-1] == '1':
                #正样本
                imgs_path.append(TRAIN_DATA_PATH+line_split[0]+'.jpg')
    
        return imgs_path
    
```

```python
#20个SVM分类
class SVM_set(object):
    def __init__(self):
        self.classes_num = CLASSES_NUM
        
        self.img_generator = Img_generator() #IoU
        self.object_load = Object_load()
        self.alexnet = Alexnet_finetune()
        
        self.svms = [] #svm集合 20个类别 就有20个svm
        
        #防止显存溢出添加
        self.resource_threshold_shape_0 = 800 #候选框超过800 将其拆分为两块处理 #1000还是有点大
        
        #self.train_all_svm() #自调
        
    
    def train_svm_with_label(self , label):
        if os.path.exists('../RCNN/qp/svm_model/svm_%d.m' % label):
            #模型已经存在 载入即可
            print('exist,loading......')
            self.svms.append(joblib.load('../RCNN/qp/svm_model/svm_%d.m' % label))
            
            print('finish loading')
            return 
        
        imgs_path = self.object_load.load(label)
        
        svm = SVC(probability = True , class_weight='balanced') #为True能输出类别的概率值 自动平衡训练数据
        
        #没有model的保存数据就先训练
        if not os.path.exists('../RCNN/qp/finetune_alexnet/checkpoint'):
            self.alexnet.train() 
            
        with tf.Session() as sess:
            sess.run(tf.global_variables_initializer())
            self.alexnet.saver.restore(sess , tf.train.latest_checkpoint('../RCNN/qp/finetune_alexnet/')) #读取模型整个model的模型参数

            scores = 0.0 #debug
            count = 0 #debug
            
            images_count = len(imgs_path) #debug
            step = 0 #debug
            for path in imgs_path:
                print('No.%d/%d' % (step , images_count)) #debug
                step = step+1 #debug
                
                #每次使用label类型的一张图片进行训练
                #将所有的label对应的图片全部训练完
                regions , labels , boxes , G_boxes = self.img_generator.one_img_rect_region_label_svm(path , label)
                
                if np.sum(labels) == label*len(labels) or np.sum(labels) == 0:
                    #说明所有的label都一样 那就不训练了 而且fit函数报错
                    continue
                
                #显存太小 不得不增加判断逻辑 这会增加运行时间
                if regions.shape[0] > self.resource_threshold_shape_0:
                    #拆开运行训练
                    for idx in range(regions.shape[0] // self.resource_threshold_shape_0):
                        labels_sum = np.sum(labels[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])
                        if labels_sum == label*self.resource_threshold_shape_0 or labels_sum == 0:
                            continue
                        
                        regions_cnn_features_block = sess.run(self.alexnet.features , feed_dict={self.alexnet.x : regions[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0] , self.alexnet.keep_prob:1.0})
                        svm.fit(regions_cnn_features_block , labels[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])

                        scores = scores + svm.score(regions_cnn_features_block , labels[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])
                        count = count+1
                    
                    #余下的部分（不是整的）
                    labels_sum = np.sum(labels[(idx+1)*self.resource_threshold_shape_0:])
                    if labels_sum == label*len(labels[(idx+1)*self.resource_threshold_shape_0:]) or labels_sum == 0:
                        continue
                            
                    regions_cnn_features_extra = sess.run(self.alexnet.features , feed_dict={self.alexnet.x : regions[(idx+1)*self.resource_threshold_shape_0:] , self.alexnet.keep_prob:1.0})
                    svm.fit(regions_cnn_features_extra , labels[(idx+1)*self.resource_threshold_shape_0:])
                    
                    scores = scores + svm.score(regions_cnn_features_extra , labels[(idx+1)*self.resource_threshold_shape_0:])
                    print(scores) #debug
                    count = count+1
                
                else:
                    regions_cnn_features = sess.run(self.alexnet.features , feed_dict={self.alexnet.x : regions , self.alexnet.keep_prob:1.0})
                    svm.fit(regions_cnn_features , labels)

                    scores = scores + svm.score(regions_cnn_features , labels)
                    print(scores) #debug
                    count = count+1
        
        try:
            print('label:%s average_score:%f' % (LABEL2STR[label] , scores/count)) #debug
        except ZeroDivisionError:
            pass #这里不会被执行到

        #先存到变量中 再保存至磁盘
        self.svms.append(svm)
        joblib.dump(svm , '../RCNN/qp/svm_model/svm_%d.m' % label) #将svm保存起来

    
    def train_all_svm(self):
        svms_path = glob('../RCNN/qp/svm_model/*.m')
        if len(svms_path) == len(LABEL2STR) - 1:
            #存在已经训练好的模型 就不再训练了 直接读取训练好的模型即可
            print('loading all model......')
            for label in range(1 , len(LABEL2STR)):
                path = '../RCNN/qp/svm_model\svm_%d.m' % label
                print('loading %s model ......' % path)
                self.svms.append(joblib.load(path))
                
            print('finish loading')
            return 
        
        for label in range(1 , len(LABEL2STR)):
            self.train_svm_with_label(label)

    
    def __meta_predict(self , regions_cnn_features , svm , rects): #proposal*4096
        labels = svm.predict(regions_cnn_features) #预言的labels 含有背景和object
        
        #print(labels) #debug
        
        probability = svm.predict_proba(regions_cnn_features) #概率信息 为NMS做准备
        
        #print(probability) #debug
        
        rects_hat = [] #label对应不是背景的标记框 才予以显示和bbox回归 label是背景的候选框不需要进行bbox回归
        probability_hat = [] #label对应不是背景的概率值 NMS score需要
        
        for i in range(len(labels)):
            if labels[i] != 0: #不是背景
                rects_hat.append(rects[i])
                probability_hat.append(np.max(probability[i]))
        
        rects_hat = np.array(rects_hat)
        probability_hat = np.array(probability_hat)
        
        if len(probability_hat) != 0:
            sorted_idx = np.argsort(probability_hat) #对概率进行升序排列 NMS score使用
            #框坐标和概率score都升序
            rects_hat = rects_hat[sorted_idx][::-1] #降序
            probability_hat = probability_hat[sorted_idx][::-1] #降序
        
        return rects_hat , probability_hat #返回非背景的候选框 每个候选框的概率值 对这些概率值进行argsort得到index 从而也将rects_hat排序了 进行后续NMS

    def nms(self , rects_hat , probability_hat): #已降序处理
        #非极大值抑制 #在同一个label之间进行抑制
        final_rects_hat = []
        final_probability_hat = [] #应该用不上了
        
        length = len(probability_hat)
        lost_flag = [1]*length #标记丢弃的框
        
        max_score_idx = 0 #记录当前最大score的idx
        
        while max_score_idx < length:
            max_score_rect = rects_hat[max_score_idx]
            
            for i in range(max_score_idx+1 , length): #rects_hat[max_score_idx:]:
                if lost_flag[i] == 1 and self.img_generator.IoU( max_score_rect , rects_hat[i] ) > NMS_IoU_THRESHOLD: #大于阈值 丢弃
                    lost_flag[i] = 0

            max_score_idx_bak = max_score_idx #后续使用
            
            #让max_score_idx指向下一个没被丢弃的最大值
            for i in range(max_score_idx+1 , length):
                if lost_flag[i] == 1:
                    max_score_idx = i
                    break
            
            #说明max_score_idx没有移动过 即后续的都被丢弃了
            if max_score_idx == max_score_idx_bak:
                break
        
        for i in range(length):
            if lost_flag[i] == 1:
                final_rects_hat.append(rects_hat[i])
                final_probability_hat.append(probability_hat[i])
                
        return np.array(final_rects_hat) , np.array(final_probability_hat)
    
    def predict(self , path):
        img_arr , rects , regions = self.img_generator.one_img_rect_region(path) #原图像数据 候选框坐标 候选框数据
        regions_cnn_features = self.alexnet.extract_feature(regions)
        
        all_rects_hat_and_proba_and_labels = [] #框坐标 标记 概率值
        
        #使用所有的svm
        for label in range(1 , len(LABEL2STR)): #20
            rects_hat , probability_hat = self.__meta_predict(regions_cnn_features , self.svms[label-1] , rects)
            
            nms_rects_hat , nms_probability_hat = self.nms(rects_hat , probability_hat)
            
            all_rects_hat_and_proba_and_labels.append( (nms_rects_hat , nms_probability_hat , label ) ) #每一个元素是框子坐标和label
            
        return all_rects_hat_and_proba_and_labels , img_arr
```

```python
#bounding-box回归
class Bbox_regression(object):
    def __init__(self):
        self.ridges = [] #pascal voc 20类别就使用20个线性回归
        #=========
        self.one_ridge = Ridge() #所有的类别只使用1个线性回归
        #=========
        self.object_load = Object_load()
        self.img_generator = Img_generator()
        self.svm_set = SVM_set()
        #self.alexnet = Alexnet_finetune()
        self.pr_generator = Clip_img()
        
        self.resource_threshold_shape_0 = 800 #同svm类
        
        ##self.train_all_ridge() #自调
        self.train_one_ridge() #自调
        
    def train_ridge_with_label(self , label):
        ##if os.path.exists('../RCNN/qp/linear_model/ridge_%d.m' % label):
        ##    #模型已经存在 载入即可
        ##    print('exist,loading......')
        ##    self.ridges.append(joblib.load('../RCNN/qp/linear_model/ridge_%d.m' % label))
        ##    
        ##    print('finish loading')
        ##    return 
        
        imgs_path = self.object_load.load(label)
        ##ridge = Ridge(alpha=1000) #正则参数为1000
        
        #没有model的保存数据就先训练
        if not os.path.exists('../RCNN/qp/finetune_alexnet/checkpoint'):
            self.svm_set.alexnet.train() 
            
        with tf.Session() as sess:
            sess.run(tf.global_variables_initializer())
            self.svm_set.alexnet.saver.restore(sess , tf.train.latest_checkpoint('../RCNN/qp/finetune_alexnet/')) #读取模型整个model的模型参数

            scores = 0.0 #debug
            count = 0 #debug
            
            images_count = len(imgs_path) #debug
            step = 0 #debug
            for path in imgs_path:
                print('No.%d/%d' % (step , images_count)) #debug
                step = step+1 #debug
                
                regions , targets = self.img_generator.one_img_rect_region_label_bnd(path , label)
                
                if len(regions) == 0:
                    #没有数据
                    continue
                
                #显存太小 不得不增加判断逻辑 这会增加运行时间
                if regions.shape[0] > self.resource_threshold_shape_0:
                    #拆开运行训练
                    for idx in range(regions.shape[0] // self.resource_threshold_shape_0):
                        regions_cnn_features_block = sess.run(self.svm_set.alexnet.features , feed_dict={self.svm_set.alexnet.x : regions[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0] , self.svm_set.alexnet.keep_prob:1.0})
                        ##ridge.fit(regions_cnn_features_block , targets[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])
                        self.one_ridge.fit(regions_cnn_features_block , targets[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])

                        ##scores = scores + ridge.score(regions_cnn_features_block , targets[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])
                        scores = scores + self.one_ridge.score(regions_cnn_features_block , targets[idx*self.resource_threshold_shape_0 : (idx+1)*self.resource_threshold_shape_0])
                        count = count+1
                    
                    #余下的部分（不是整的） 
                    regions_cnn_features_extra = sess.run(self.svm_set.alexnet.features , feed_dict={self.svm_set.alexnet.x : regions[(idx+1)*self.resource_threshold_shape_0:] , self.svm_set.alexnet.keep_prob:1.0})
                    ##ridge.fit(regions_cnn_features_extra , targets[(idx+1)*self.resource_threshold_shape_0:])
                    self.one_ridge.fit(regions_cnn_features_extra , targets[(idx+1)*self.resource_threshold_shape_0:])
                    
                    ##scores = scores + ridge.score(regions_cnn_features_extra , targets[(idx+1)*self.resource_threshold_shape_0:])
                    scores = scores + self.one_ridge.score(regions_cnn_features_extra , targets[(idx+1)*self.resource_threshold_shape_0:])
                    count = count+1
                
                else:
                    regions_cnn_features = sess.run(self.svm_set.alexnet.features , feed_dict={self.svm_set.alexnet.x : regions , self.svm_set.alexnet.keep_prob:1.0})
                    ##ridge.fit(regions_cnn_features , targets)
                    self.one_ridge.fit(regions_cnn_features , targets)

                    ##scores = scores + ridge.score(regions_cnn_features , targets)
                    scores = scores + self.one_ridge.score(regions_cnn_features , targets)

                    count = count+1
                    
        try:
            print('label:%s average_score:%f' % (LABEL2STR[label] , scores/count)) #debug
        except ZeroDivisionError:
            pass #这里不会被执行到    
        
        ##self.ridges.append(ridge)
        ##joblib.dump(ridge , '../RCNN/qp/linear_model/ridge_%d.m' % label) #将ridge保存起来
    
    def train_all_ridge(self):
        ridges_path = glob('../RCNN/qp/linear_model/*.m')
        
        if len(ridges_path) == len(LABEL2STR) - 1:
            #存在已经训练好的模型 就不再训练了 直接读取训练好的模型即可
            print('loading all model......')
            for label in range(1 , len(LABEL2STR)):
                path = '../RCNN/qp/linear_model/ridge_%d.m' % label
                print('loading %s model ......' % path)
                self.ridges.append(joblib.load(path))
                
            print('finish loading')
            return 
        
        for label in range(1 , len(LABEL2STR)):
            self.train_ridge_with_label(label)
            
    #===============
    def train_one_ridge(self):
        if os.path.exists('../RCNN/qp/one_linear_model/ridge.m'):
            #模型已经存在 载入即可
            print('exist,loading......')
            self.one_ridge = joblib.load('../RCNN/qp/one_linear_model/ridge.m')
            
            print('finish loading')
            return 
        
        for label in range(1 , len(LABEL2STR)):
            self.train_ridge_with_label(label)
        
        joblib.dump(self.one_ridge , '../RCNN/qp/one_linear_model/ridge.m') #将ridge保存起来
    #===============
        
    def __meta_predict(self , regions_cnn_features , rects , ridge):
        def to(rect):
            x1 = rect[0]
            x2 = rect[1]
            y1 = rect[2]
            y2 = rect[3]
            
            w = x2-x1
            h = y2-y1
            
            x_c = (x1+x2)//2
            y_c = (y1+y2)//2
            
            return x_c , y_c , w , h
        
        def ot(target):
            x_c = target[0]
            y_c = target[1]
            w = target[2]
            h = target[3]
            
            x1 = 0.5*(2*x_c-w)
            y1 = 0.5*(2*y_c-h)
            x2 = x1+w
            y2 = y1+h
            
            x1=int(round(x1))
            y1=int(round(y1))
            x2=int(round(x2))
            y2=int(round(y2))
            
            if x1<0:
                x1 = 0
            if x2>WIDTH:
                x2 = WIDTH
            if y1<0:
                y1 = 0
            if y2>HEIGHT:
                y2 = HEIGHT
            
        
            return [x1 , x2 , y1 , y2]
        
        def target2rect(target_hat , P_box):
            t_x = target_hat[0]
            t_y = target_hat[1]
            t_w = target_hat[2]
            t_h = target_hat[3]
            
            P_x , P_y , P_w , P_h = to(P_box)
            
            G_x_hat = P_w*t_x+P_x
            G_y_hat = P_h*t_y+P_y
            G_w_hat = P_w*np.exp(t_w)
            G_h_hat = P_h*np.exp(t_h)
            
            return ot([G_x_hat , G_y_hat , G_w_hat , G_h_hat]) #ot还需要转化为(x1,x2,y1,y2)形式
            
        target_hat = ridge.predict(regions_cnn_features)
        rect_hat = []
        
        for i in range(len(target_hat)):
            rect_hat.append(target2rect(target_hat[i] , rects[i]))
            
        return np.array(rect_hat)
    
    
    def predict(self , path):
        
        all_rects_hat_and_proba_and_labels , img_arr=self.svm_set.predict(path)
        print('finish svm') #debug
        bbox_r_rects_hat_and_label = []
        
        for (nms_rects_hat , _ , label) in all_rects_hat_and_proba_and_labels:
            
            if len(nms_rects_hat) == 0:
                #说明此图像不存在此label
                bbox_r_rects_hat_and_label.append((np.array([]) , label))
                continue 
            
            nms_rects_hat_region = self.pr_generator.meta_clip_region(img_arr , nms_rects_hat)
            
            nms_rects_hat_region_cnn_features = self.svm_set.alexnet.extract_feature(nms_rects_hat_region)
    
            ##nms_rects_hat_bb_r = self.__meta_predict(nms_rects_hat_region_cnn_features , nms_rects_hat , self.ridges[label-1])
            nms_rects_hat_bb_r = self.__meta_predict(nms_rects_hat_region_cnn_features , nms_rects_hat , self.one_ridge)

        
            bbox_r_rects_hat_and_label.append((nms_rects_hat_bb_r , label))
    
        return bbox_r_rects_hat_and_label , img_arr #最终的框
```

```python
#整合所有的模块
class RCNN(object):
    def __init__(self):        
        self.process_m = Process_img() #图像加载模块
        self.clip_m = Clip_img() #图像候选区域产生模块
        self.extract_m = Extract_img() #候选区域特征提取模块
        self.classify_m = Classify_img() #类别分类模块
        self.regression_m = Regression_bbox() #bb盒回归模块
        
        self.display_m = Display() #可视化模块
        
        self.epochs = EPOCHS
        self.iter = self.epochs // BATCH_SIZE #一个epoch共多少次
    
    def train(self):
        
        for i in range(self.epochs):
            for j in range(self.iter):
                #原图像 候选区域具体数据 候选区域的人工标记 候选区域的位置信息 图像的名字
                img , proposals , labels , G_box , img_names = self.process_m.load() #读batch_size个图像 一般1个
                
                #候选区域已经产生 shape: batch_size*proposal*4
                img_clip = self.clip_m.clip(img)
                
                img_clip_feature = self.extract_m.extract(img_clip) #shape: batch_size*proposal*4096
                
                proposals_feature = self.extract_m.extract(proposals) #shape: batch_size*proposal*4096
                
                #batch_size*proposal*4096 batch_suze*proposal*1
                self.classify_m.train(proposals_feature , labels) #训练svm
                
                #bbox回归
                #shape: batch_size*proposal*4 \ shape: batch_size*proposal*4
                self.regression_m.train(img_clip , G_box) #一个是区域选择算法产生的 一个是数据集中的人工标注的
                
    
    def predict(self):
        img = self.process_m.load_test() #读一张测试图像
        
        img_clip = self.clip_m.clip(img) #batch_size*proposal*4 #候选区域的具体位置
        
        img_clip_region = self.clip_m.clip_region(img) #batch_size*proposal*height*width*3 #候选区域的具体数据
        
        img_clip_region_feature = self.extract_m.extract(img_clip_region) #batch_size*propposal*4096
        
        #下面都是预测的函数
        #预测类别
        img_labels_pred = self.classify_m.predict(img_clip_region_feature)
        
        #框微调
        img_G_box_pred = self.regression_m.predict(img_clip) #batch_size*proposal*4
        
        #可视化
        self.display_m.display(img , img_labels_pred , img_G_box_pred)
```

```python
#alexnet调优demo
#demo
class Image_generator(object):
    def __init__(self , image_path_train = '../RCNN/images/'):
        self.IMAGE_PATH_TRAIN = image_path_train
        
        #self.LABEL_PATH_TRAIN = '../../tensorflow2/dataset/CIFAR10/trainLabels.csv'
        
        self.LABEL2INDEX = {'cat':0 , 'dog':1}
        self.INDEX2LABEL = {value:key for key , value in self.LABEL2INDEX.items()}
        
        self.CLASSES_NUM = 2 #dog vs cat

        #self.train_labels = pd.read_csv(self.LABEL_PATH_TRAIN)
        #self.train_labels = self.train_labels['label'].map(self.LABEL2INDEX).get_values()
        
        self.image_mean = np.array([104. , 117. , 124.])# , dtype=np.float32)
        
        self.image_names = glob(self.IMAGE_PATH_TRAIN + '*') #所有图片的路径
        
        self.image_num = len(self.image_names) #训练集大小
        
        self.train_image_num = int(0.9*self.image_num) #训练集大小
        self.val_image_num = int(0.1*self.image_num) #验证集大小
      
    def get_image_num(self):
        return self.train_image_num , self.val_image_num #返回大小

    def next_batch(self , batch_size , is_training = True):
        if is_training:
            #训练数据集
            idx = np.random.randint(low = 0 , high = self.train_image_num , size = batch_size) #随机索引
        else:
            #验证数据集
            idx = np.random.randint(low = self.train_image_num , high=self.train_image_num+self.val_image_num , size = batch_size) #随机索引
        
        images = []
        labels = []
        
        for i in idx:
            #使用imageio.imread打开出现错误 卡机 迁移失败
            #img = cv2.imread(self.IMAGE_PATH_TRAIN + '%d' % (i+1) + '.png')
            img = cv2.imread(self.image_names[i]) #BGR形式 height*width*channel样式
            
            #img = img[: , : , [2,1,0]] #BGR还原为RGB
            
            #print(self.image_names[i]) #debug
            
            img = cv2.resize(img , dsize=(HEIGHT , WIDTH)) #resize 适合alexnet处理
            #print(i , img.shape) #debug
            #img.astype(np.float32)
            
            img = img - self.image_mean #减去图像的平均像素值
            
            images.append(img)
            
            #print(self.INDEX2LABEL[self.train_labels[i]]) #debug
            labels.append( self.LABEL2INDEX[self.image_names[i].split('\\')[1].split('.')[0]] )
            #labels.append(self.train_labels[i])
        
        return np.array(images) , to_categorical(labels , num_classes=self.CLASSES_NUM)

#================================
#================================
#alexnet调优 demo
#demo

#refer:https://blog.csdn.net/two_vv/article/details/76769860

class Alexnet_prime(object):
    def __init__(self , dataset_path = '../RCNN/images/' , classes_num=2):
        self.batch_size = 128
        self.learning_rate = 0.001
        self.num_epochs = 10
        self.dropout_rate = 0.5
        self.train_layers = ['fc7' , 'fc8'] #只训练fc7 fc8层 别的层使用预训练参数
        
        #=====
        self.classes_num = classes_num
        #=====
        
        #模型可视化 模型保存相关
        #==========
        self.filewriter_path = 'qp/image'
        self.checkpoint_path = 'qp/finetune_alexnet'
        #==========
        
        #==========
        self.generator = Image_generator() #图片产生器
        self.train_image_num , self.val_image_num = self.generator.get_image_num()
        #==========
        
        #下面是tensorflow部分
        self.x = tf.placeholder(tf.float32 , shape=[None , HEIGHT , WIDTH , CHANNEL])
        self.y = tf.placeholder(tf.float32 , shape=[None , self.classes_num]) #None 行不定 tf自行完成判定
        self.keep_prob = tf.placeholder(tf.float32) #scalar
        
        self.model = AlexNet_model(self.x , self.keep_prob , self.classes_num , self.train_layers)

        self.score = self.model.fc8
    
        #需要自己训练的层参数
        #fc7 weights biases
        #fc8 weights biases
        self.var_list = [v for v in tf.trainable_variables() if v.name.split('/')[0] in self.train_layers]
        
        #损失函数 交叉熵 softmax
        with tf.name_scope('cross_ent'):
            self.loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=self.score , labels=self.y))
        
        #构造训练过程
        with tf.name_scope('train'):
            #损失对fc7 weights biases fc8 weights biases的梯度
            self.gradients = tf.gradients(self.loss , self.var_list) #损失对训练参数的梯度
            self.gradients = list(zip(self.gradients , self.var_list))
            
            #没有使用minimize 而是使用appl_gradients方式 梯度下降
            self.optimizer = tf.train.GradientDescentOptimizer(self.learning_rate)
            self.train_op = self.optimizer.apply_gradients(grads_and_vars = self.gradients)
            
        #可视化使用
        for gradient , var in self.gradients:
            tf.summary.histogram(var.name + '/gradient' , gradient)
        
        for var in self.var_list:
            tf.summary.histogram(var.name , var)
            
        tf.summary.scalar('cross_entropy' , self.loss)
        
        #精度
        with tf.name_scope('accuracy'):
            self.corrent_pred = tf.equal(tf.argmax(self.score , axis=1) , tf.argmax(self.y , axis=1)) #返回的是最大值的索引
            self.accuracy = tf.reduce_mean(tf.cast(self.corrent_pred , tf.float32))
        
        tf.summary.scalar('accuracy' , self.accuracy)
        
        #merge所有summary 使用tensorboard显示
        self.merged_summary = tf.summary.merge_all() #需要有session时才能起作用
        
        #tensorboard使用 日志形式 在此文件夹下运行tensorboard
        self.writer = tf.summary.FileWriter(logdir = self.filewriter_path ) #qp/image
        
        #将模型保存起来 持久模式 以后能够读出使用
        self.saver = tf.train.Saver()
         
    def train(self):
        
        with tf.Session() as sess:
            sess.run(tf.global_variables_initializer())
            
            #计算图放到tensorboard中
            self.writer.add_graph(sess.graph)
            
            self.model.load_parameter(sess)
            
            for i in range(self.num_epochs):
                #一个epoch
                
                step = 1 #tensorboard使用
                iters = self.train_image_num // self.batch_size
                
                #training step
                for j in range(2):
                #for j in range(iters):
                    train_images , train_labels = self.generator.next_batch(self.batch_size)
                    
                    loss , _ = sess.run([self.loss , self.train_op] , feed_dict={self.x:train_images , self.y:train_labels , self.keep_prob:self.dropout_rate })
                    
                    #save model
                    #生成训练时的相关数据 上面使用tf.summary.xxx函数的数据
                    merge_summary = sess.run(self.merged_summary , {self.x:train_images , self.y:train_labels , self.keep_prob:1.0 })
                    self.writer.add_summary(merge_summary , i*iters+j) #写入文件
                    
                    step = step + 1
                    
                    print('epoch:%d/%d iter:%d/%d loss:%f' % (i , self.num_epochs , j , iters , loss))

                #val step
                accu_i = 0.0
                accu_c = 0
                for k in range(2):
                #for k in range(self.val_image_num // self.batch_size):
                    val_images , val_labels = self.generator.next_batch(self.batch_size , is_training = False)
    
                    _ , loss = sess.run([self.accuracy , self.loss] , feed_dict={self.x:val_images , self.y:val_labels , self.keep_prob:1.0 })
                    print('accuracy_i:%f loss:%f' % (_ , loss))
            
                    accu_i = accu_i + _
                    accu_c = accu_c + 1
                
                accu = accu_i / accu_c

                print('val accuracy:%f' % accu)
                
                checkpoint_name = os.path.join(self.checkpoint_path , 'model_epoch'+str(i)+'.ckpt')
                save_path = self.saver.save(sess , checkpoint_name)
                
        self.writer.close()

#
#alexnet = Alexnet_prime()
#alexnet.train()

#alexnet复现以及预训练参数导入demo
#使用网络下载的图片进行测试

x=tf.placeholder(tf.float32 , [5,227,227,3]) #一次测试一张图片

keep_prob = tf.placeholder(tf.float32)

model = AlexNet_model(x,keep_prob ,num_classes=1000 , skip_layer=[])

score=model.fc8

softmax=tf.nn.softmax(score)
#上面代码片段只运行一次
#==========

g = Image_generator(image_path_train='../RCNN/cat_and_dog/')
imgs , _ = g.next_batch(batch_size=11)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    model.load_parameter(sess)
    
    probs=sess.run(softmax  ,feed_dict={x:imgs , keep_prob:1.0})

    for i in range(probs.shape[0]):
        print(class_names[np.argmax(probs[i])])
    
```
